```python
#Рабочий вариант
import cv2 as cv
from PIL import Image

#Открывает видео
cap = cv.VideoCapture(r"C:\Users\mrafa\Desktop\IMG_7259.MP4")
#Проверяет первый кадр
ok, firstFrame = cap.read()
#Проверяем ошибку
if not ok or firstFrame is None:
    raise SystemExit

#список для сохранённых кадров
imgs = []

# Координаты области 
x1 = 100
y1 = 50
x2 = 700
y2 = 400

# Проверка координат
h, w = firstFrame.shape[:2]
if not (0 <= x1 < x2 <= w and 0 <= y1 < y2 <= h):
    raise ValueError("Неверные координаты:", w, h)

# Показываем первый кадр с рамкой
preview = firstFrame.copy()
cv.rectangle(preview, (x1, y1), (x2, y2), (0, 255, 0), 2)
cv.imshow("", preview)

print("Проверь рамку")
print("Нажми любую клавишу, чтобы запустить")
print("Нажми q, чтобы отменить запуск")

key = cv.waitKey(0)

# Если нажали q — отменяем запуск
if key == ord("q"):
    cv.destroyAllWindows()
    cap.release()
    print("Запуск отменён")
    


while True:
    #Продолжает работу с другими кадрами
    ok, frame = cap.read()
    if frame is None:
        break
        
    #Вырезает указанную область
    slide = frame[y1:y2, x1:x2]
    #Показывает вырезанный участок
    cv.imshow("", slide)
    k = cv.waitKey(30)

    # если прошло 2 секунды — сохраняем кадр
    if k == ord("s"):
        #Сохраняет текущий кадр
        imgs.append(Image.fromarray(cv.cvtColor(slide, cv.COLOR_BGR2RGB)))
        print("Сохранено кадров:", len(imgs))
    
    #Останавливает цикл
    if k == ord("q"):
        break

#Закрывает все окна
cv.destroyAllWindows()
cap.release()

#сохраняет пдф
if imgs:
    imgs[0].save("slides.pdf", save_all=True, append_images=imgs[1:])
    print("Готово: slides.pdf, страниц:", len(imgs))

else:
    print("Ни одного кадра не сохранено") 
```
